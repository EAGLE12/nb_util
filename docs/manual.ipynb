# 前置き

jupyte noteookは，json(JavaScript Object Notation)で書かれています．
jsonというのは「データを表現するための記法（≒文法）」です．
人間にデータを説明する場合は，「注意書き」や「表」を駆使してデータを伝えられるが，
相手がコンピュータの場合，厳格なルールに沿って伝えたい内容を記述しなければなりません．
その記述ルールの一つとしてjsonであり，jupyter notebookで採用されています．

# jupyter notebookの中身
jupyter notebookの中身を確認するときに，わざわざjupyte notebookを起動して確認するのは面倒です．
そこで，ファイルの中身を確認する便利なコマンドがあります．それは，catコマンドです.

`cat test.ipynb`

実行してみると以下のようになります．

しかし，jupyter notebookは実行内容とは別に，様々な情報が明記されています．
そのため，そのままcatで表示すると，書いた内容だけでなく，
必要でない情報まで表示されてしまいます．

そこで，内容だけを表示するプログラムが，puts_ipynb.rbです．

`ruby puts_ipynb.rb`

# 二つのjupyter nortebookを一つに
もし，二つのjupyte notebookのファイルを結合する場合，わざわざ二つのファイルを開いて，
コピーしてペーストする方法が，第一に考えます．
しかし，それはかなり面倒な作業です．
しかし，join_ipynb.rbを使用すると，二つのファイルを結合して，1つのipynbにまとめることができます．
この機能は，例えば，個人がまとめた複数のファイルをまとめてメンバーに配布する際に便利です．

二つのファイルを1つのjupyter notebookに結合する方法：

`ruby join_ipynb.rb file1.ipynb file2.ipynb`

tmp.ipynb ファイルが生成されます．





# jupyter notebookはどこ？
プロジェクトを始めたばかりだとか，課題に取り組みたての時は，
階層が浅かったり，まだどこに自分が欲しいファイルがあるかは，鮮明に覚えているはずです．
しかし，階層が深くなり，どんどん作業ファイルが増えてきた時，自分がどこに何を置いたかわからなくなります．
特に最近は，”最近使用した履歴”という補助機能のあるアプリケーションが少なくありません．
そのため，実態のファイルがどこにあるのか，迷子に陥ってしまします．こうなれば，後は終わりです．．．
そーなったときに使用するのが，treeコマンドではないでしょうか．

`tree`

一覧が表示されるが，色は単色で，jupyter notebookはどこにあるか，
パッと見ただけでは，なかなかわかりません．そこで，.ipynbファイルだけを水色に，他を緑に変えて表示するプログラムが，
rtree.rbです．

`ruby rtree.rb`

# jupyterでruby

jupyterでrubyを実行するためには，rubyを実行するkernel環境で，scirubyによって開発されているirubyを入れておく必要があります[iruby](https://github.com/SciRuby/iruby)．

rubyのgemみたいなものである，python版のpipであらかじめ，
jupyterとipythonが入っているとします．

OSXにパッケージをinstallするbrewを使います．以下をterminalで実行
```
brew install zeromq czmq
gem install rbczmq -- --with-system-libs
gem install iruby
```

zeroMQはコードを実行する中間layerを提供するlibraryのようです[zeromq](http://zeromq.org/intro:read-the-manual)．

* (備考)なお、Macならrbczmqの代わりにffi-rzmq、とも言われている...
* [参考](http://qiita.com/domitry/items/aa348caf993e876a7be5)

```
#環境によってffi-rzmqでは動作しない場合があるかもなので...
gem install ffi-rzmq
gem install iruby
```

立ち上がりはするのですが，すぐにDead kernelとなって正常に動作しない場合があります．
結局，gem uninstall ffi-rzmqして上のやり方でrbczmqを入れると解決するかも．
